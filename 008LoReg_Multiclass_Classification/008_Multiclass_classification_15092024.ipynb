```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
irish_df = pd.read_csv('/content/drive/MyDrive/Data Science/(Week_13_14_15_16_17) Machine Learning (ML)/CSV /IRIS.csv')
irish_df
```

```python
irish_df['species'].unique()
```

```python
irish_df['species'].replace({'Iris-setosa':1, 'Iris-versicolor':2, 'Iris-virginica':3}, inplace=True)
irish_df
```

```python
from sklearn.model_selection import train_test_split
X = irish_df[['sepal_length', 'sepal_width','petal_length','petal_width']]
y = irish_df[['species']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.predict(X_test) #'Iris-setosa':1, 'Iris-versicolor':2, 'Iris-virginica':3
```

```python
model.score(X_test, y_test)
```
