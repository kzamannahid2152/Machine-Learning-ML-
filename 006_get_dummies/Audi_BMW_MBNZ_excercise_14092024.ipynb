```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# import warnings
# warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Data Science/(Week_13_14_15_16_17) Machine Learning (ML)/CSV /BMW_carprices (1).csv')
df
```

```python
dummies  = pd.get_dummies(df['Car Model']).astype(int)
dummies
```

```python
two_table = pd.concat([df,dummies],axis='columns')
two_table
```

```python
newdf= two_table.drop(['Car Model','BMW X5'],axis='columns')
newdf
```

```python
X = newdf.drop(['Sell Price($)'],axis='columns')
X
```

```python
y = newdf['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
model.score(X,y)
```
