```python
df = pd.read_csv('/content/drive/MyDrive/Data Science/(Week_13_14_15_16_17) Machine Learning (ML)/CSV /homeprices (1).csv')
df
```

**Converting into DUMMY variable using___"get_dummies"___command**

```python
dummies = pd.get_dummies(df.town).astype(int)
dummies
```

**Converting two table with togather using_________"pd.concat"___________**

```python
df_concat_table = pd.concat([df, dummies], axis='columns')
df_concat_table
```

**Deleting "town" and "west windor" column**

```python
df_concat_table
```

```python
df_concat_table.drop(['town', 'west windsor'], axis='columns', inplace=True)
df_concat_table
```

**Deleting Price column also because this is our target (dependent_value). we have predict this build a model**

```python
df_concat_table.drop(['price'], axis='columns', inplace=True)
df_concat_table
```

```python
X = df_concat_table
X
```

```python
y = df['price']
y
```

**Building LOGISTIC REGRESSION MODEL**

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X, y) #training model
```

```python
model.predict([[2800,0,0]])
```
